```python
def process_salary(salary_str):
    if pd.isna(salary_str) or salary_str == "":
        return np.nan 
    clean_str = salary_str.lower().replace('$', '').replace(',', '').replace(' ', '')

    if 'k' in clean_str:
        clean_str = clean_str.replace('k', '')
        multiplier = 1000
    else:
        multiplier = 1

    try:
        if '-' in clean_str:
            low, high = clean_str.split('-')
            avg = (float(low) + float(high)) / 2
            return avg * multiplier
        else:
            return float(clean_str) * multiplier
    except ValueError:
        return np.nan
```

```python
df['salary_Fixed'] = df['salary_range'].apply(process_salary)
```

```python
import re
def removehtml(text):
    if pd.isna(text):
        return ""
    text = re.sub(r'<[^>]+>', ' ', str(text))
    
    text = text.replace('\n', ' ').replace('\t', ' ').replace('\r', ' ')
    
    text = text.replace('/n', ' ').replace('\\n', ' ')
    
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python
import spacy
import numpy as np
import pandas as pd


nlp = spacy.load("en_core_web_md")

print("Converting text to vectors... This takes a moment.")


def get_vec(text):
    return nlp(str(text)).vector

text_train_vectors = np.vstack(text_train.apply(get_vec))
text_test_vectors = np.vstack(text_test.apply(get_vec))

print("Vectorization Complete! New shape:", text_train_vectors.shape)
```

```python
from scipy.sparse import hstack
from sklearn.ensemble import VotingClassifier, RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix

X_train_final_vec = np.hstack([X_train_scaled, text_train_vectors])
X_test_final_vec = np.hstack([X_test_scaled, text_test_vectors])


rf_expert = RandomForestClassifier(n_estimators=200, 
                                   class_weight='balanced_subsample', 
                                   random_state=42)

xgb_expert = XGBClassifier(use_label_encoder=False, 
                           eval_metric='logloss',
                           scale_pos_weight=5, 
                           max_depth=4,
                           learning_rate=0.05,
                           n_estimators=200,
                           random_state=42)

voting_model = VotingClassifier(estimators=[('rf', rf_expert), ('xgb', xgb_expert)], 
                                voting='soft')

print("Training on Semantic Vectors...")
voting_model.fit(X_train_final_vec, y_train)

y_pred_vec = voting_model.predict(X_test_final_vec)

print("\n--- Results with Spacy Embeddings ---")
print(classification_report(y_test, y_pred_vec))
print(confusion_matrix(y_test, y_pred_vec))
```
