```python
def process_salary(salary_str):
    if pd.isna(salary_str) or salary_str == "":
        return np.nan 
    clean_str = salary_str.lower().replace('$', '').replace(',', '').replace(' ', '')

    if 'k' in clean_str:
        clean_str = clean_str.replace('k', '')
        multiplier = 1000
    else:
        multiplier = 1

    try:
        if '-' in clean_str:
            low, high = clean_str.split('-')
            avg = (float(low) + float(high)) / 2
            return avg * multiplier
        else:
            return float(clean_str) * multiplier
    except ValueError:
        return np.nan
```

```python
df['salary_Fixed'] = df['salary_range'].apply(process_salary)
```

```python
import re
def removehtml(text):
    if pd.isna(text):
        return ""
    text = re.sub(r'<[^>]+>', ' ', str(text))
    
    text = text.replace('\n', ' ').replace('\t', ' ').replace('\r', ' ')
    
    text = text.replace('/n', ' ').replace('\\n', ' ')
    
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python
df['final_text'] = df['final_text'].apply(removehtml)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
X = df.drop(['fraudulent', 'final_text'], axis=1)
y = df['fraudulent']
```

```python
X_train, X_test, y_train, y_test, text_train, text_test = train_test_split(
    X, y, df['final_text'], test_size=0.2, random_state=42, stratify=y
)
```

```python

tfidf = TfidfVectorizer(max_features=5000)
```

```python
text_train_tfidf = tfidf.fit_transform(text_train)
text_test_tfidf = tfidf.transform(text_test)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from scipy.sparse import hstack
x_train_final = hstack([X_train_scaled, text_train_tfidf])
x_test_final = hstack([X_test_scaled, text_test_tfidf])
```

```python
log_reg = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
```

```python
print("Training model...")
log_reg.fit(x_train_final, y_train)
```

```python
y_pred = log_reg.predict(x_test_final)
```

```python
print("\n--- Model Performance ---")
print(classification_report(y_test, y_pred))

print("\n--- Confusion Matrix ---")
print(confusion_matrix(y_test, y_pred))
```
